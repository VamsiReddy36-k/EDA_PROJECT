# Phase 2: Data Cleaning & Manipulation

## Handle Missing Data

```python
data.isnull().sum()
```

### Observation 

* There are **missing values in seven columns**, mainly related to **social media usage, productivity, stress, and sleep**.
* The missing counts range from around **1,600 to 2,700 entries** per column.
* Most other columns are **complete and consistent**, with no null values.
* Overall, the dataset is mostly clean, requiring minor imputation for these key fields.


```python
# Select only numeric columns that have missing values
missing_cols = data.columns[data.isnull().sum() > 0]
numeric_missing_cols = data[missing_cols].select_dtypes(include=['float64', 'int64']).columns

# Set up subplot grid dynamically based on number of columns
n_cols = 2  # number of plots per row
n_rows = (len(numeric_missing_cols) + n_cols - 1) // n_cols  # calculate rows needed

plt.figure(figsize=(12, 5 * n_rows))

# Loop through numeric columns with missing values and plot KDE
for i, col in enumerate(numeric_missing_cols, 1):
    plt.subplot(n_rows, n_cols, i)
    data[col].plot(kind='kde', color='teal')
    plt.title(f'{col} Distribution', fontsize=12, weight='bold')
    plt.xlabel(col)
    plt.ylabel('Density')
    plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show() 
```

### Observation 

* The KDE plots display the **distribution of all numeric columns with missing values**.
* Most variables show **smooth and moderate distributions**, indicating consistent data patterns.
* A few columns have **slight skewness**, suggesting uneven value spread or outliers.
* Overall, the distributions appear **reasonable and suitable for imputation** using median values.


```python
# Identify numerical columns
numeric_cols = data.select_dtypes(include=[np.number]).columns

# Fill missing values with median
for col in numeric_cols:
    if data[col].isnull().sum() > 0:
        data[col].fillna(data[col].median(), inplace=True)
```

### Observation 

* Missing values in all **numerical columns** were successfully **filled with their median values**, ensuring data completeness while minimizing the impact of outliers.

## Outliers

```python
for i in data.select_dtypes(include = "number").columns:
    sns.boxplot(data = data,x = i)
    plt.show() 
```

```python
total_outliers = 0

for col in num_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = data[(data[col] < lower) | (data[col] > upper)]
    count = len(outliers)

    print(f"{col}: {count} outliers")
    total_outliers += count

print("\nTotal number of outliers across all numeric columns:", total_outliers)
```

## Observation

* The boxplots reveal that most numerical columns have **fairly consistent distributions**, though a few, such as **stress level, sleep hours, and productivity scores**, contain **some outliers** that may represent individuals with unusual behavioral or work patterns.


## Fix Inconsistencies in Categorical Data

```python
# Select categorical columns (object type)
cat_cols = data.select_dtypes(include=['object']).columns

# Clean and standardize categorical values
for col in cat_cols:
    # Convert to lowercase, strip spaces
    data[col] = data[col].astype(str).str.strip().str.lower()
    
    # Convert cleaned columns to categorical dtype
    data[col] = data[col].astype('category')
```

### Observation

All categorical columns were **cleaned and standardized** by converting text to **lowercase** and **removing extra spaces**, then changed to the **categorical data type** for consistency and better memory efficiency.
